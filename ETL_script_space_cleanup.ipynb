```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
# Load in Astronaut Dataset
file_to_load= "csv_data/astronauts.csv"
astro_df= pd.read_csv(file_to_load)
#astro_df.head()
```

```python
file_to_load= "csv_data/Space_Corrected.csv"
space_df= pd.read_csv(file_to_load)
#space_df.head(50)
```

```python
new_astro_df= astro_df[['Name','Birth Date', 'Gender', 'Status', 'Missions']].copy()
#new_astro_df.head()
```

```python
new_space_df = space_df[['Company Name','Datum', 'Detail', 'Status Mission']].copy()
new_space_df.head()
```

```python
space_data = new_space_df['Detail'].str.split("|", n=1, expand=True)

new_space_df['Missions']= space_data[1]

new_space_df.drop(columns= ['Detail'], inplace= True)

new_space_df.head(10)
```

```python
new = new_space_df['Missions'].str.split(",", n=1, expand=True)

new_space_df['Mission part 1']= new[0]
new_space_df['Mission part 2']= new[1]


new_space_df.drop(columns= ['Missions'], inplace= True)

new_space_df.head(10)
```

```python
new_2 = new_space_df['Mission part 1'].str.split("&", n=1, expand=True)

new_space_df['Mission 1']= new_2[0]
new_space_df['Mission 2']= new_2[1]

new_space_df.drop(columns= ['Mission part 1'], inplace= True)

new_space_df.head(50)
```

```python
new_3 = new_space_df['Mission part 2'].str.split("&", n=1, expand=True)

new_space_df['Mission 3']= new_3[0]
new_space_df['Mission 4']= new_3[1]

new_space_df.drop(columns= ['Mission part 2'], inplace= True)

new_space_df
```
