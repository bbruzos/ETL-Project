```python
file_to_load= "csv_data/astronauts.csv"
astro_df= pd.read_csv(file_to_load)
astro_df.head()
```

```python
file_to_load= "csv_data/science_missions.csv"
missions_df= pd.read_csv(file_to_load)
missions_df.head()
```

```python
new_astro_df= astro_df[['Name','Birth Date', 'Gender', 'Status', 'Missions']].copy()
new_astro_df.head()
```

```python
mission_data= new_astro_df['Missions'].str.split(",", n=3, expand=True)

new_astro_df['Mission 1']= mission_data[0]
new_astro_df['Mission 2']= mission_data[1]
new_astro_df['Mission 3']= mission_data[2]
new_astro_df['Mission 4']= mission_data[3]

new_astro_df.drop(columns= ['Missions'], inplace= True)

new_astro_df.head()
```
