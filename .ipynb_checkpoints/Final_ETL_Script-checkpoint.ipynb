```python
import pandas as pd
from sqlalchemy import create_engine
from config import password
```

```python
# Load in Astronaut CSV and save into DataFrame
file_to_load= "csv_data/astronauts.csv"
astro_df= pd.read_csv(file_to_load)
astro_df.head()
```

```python
#Load in Space Missions CSV and store into DataFrame
file_to_load= "csv_data/Space_Corrected.csv"
space_df= pd.read_csv(file_to_load)
space_df.head(50)
```

```python
#Extract the columns that we want from the Astronaut CSV.
new_astro_df= astro_df[['Name','Birth Date','Status','Gender','Missions']].copy()
new_astro_df.head()
```

```python
#Clean up the Data. Here, we took all of the missions in the Astronaut DF and created new columns for each mission.
#This was done by using the split method.
astro_data= new_astro_df['Missions'].str.split(",", n=3, expand=True)

new_astro_df['Mission 1']= astro_data[0]
new_astro_df['Mission 2']= astro_data[1]
new_astro_df['Mission 3']= astro_data[2]
new_astro_df['Mission 4']= astro_data[3]

new_astro_df.drop(columns= ['Missions'], inplace= True)

new_astro_df.head()
```

```python
#More clean up. We had to remove the extra parentheses in the mission columns that was needed. 
#This took out information that was not needed.
new_astro_df['Mission 1']= new_astro_df['Mission 1'].str.split('(').str[0]
new_astro_df['Mission 2']= new_astro_df['Mission 2'].str.split('(').str[0]
new_astro_df['Mission 3']= new_astro_df['Mission 3'].str.split('(').str[0]
new_astro_df['Mission 4']= new_astro_df['Mission 4'].str.split('(').str[0]
new_astro_df
```

```python
#Using the melt method, we stacked all the missions so there is one column for the missions.
transformed_astro_df= pd.melt(new_astro_df, id_vars =['Name','Birth Date','Status','Gender'], 
                            value_vars =['Mission 1','Mission 2','Mission 3','Mission 4'], 
                            var_name ='Mission_num', value_name ='Missions') 
transformed_astro_df
```

```python
#We had to further clean the data and remove any spaces before or after the mission names so it does not skew
#the join later on.
cleaned_astro_df= transformed_astro_df.dropna(how='any') 
cleaned_astro_df['Missions']= cleaned_astro_df['Missions'].str.strip()
cleaned_astro_df.head(50)
```

```python
#Extract the columns that we want from the Space Missions CSV.
new_space_df = space_df[['Company Name','Datum', 'Detail', 'Status Mission']].copy()
new_space_df.head()
```

```python
#Cleaning the Space Missions DF. The first step we took was removing the left side of the pipe in the detail column.
#These were the names of the rockets and we did not need it. We then renamed the new column to Missions and dropped
#the Detail column.
space_data = new_space_df['Detail'].str.split("|", n=1, expand=True)

new_space_df['Missions']= space_data[1]

new_space_df.drop(columns= ['Detail'], inplace= True)

new_space_df.head(10)
```

```python
#Cleaning the Missions column more. We split it by the comma. This was split into 2 columns - Mission part 1 and 
#Mission part 2.
new = new_space_df['Missions'].str.split(",", n=1, expand=True)

new_space_df['Mission part 1']= new[0]
new_space_df['Mission part 2']= new[1]


new_space_df.drop(columns= ['Missions'], inplace= True)

new_space_df.head(10)
```

```python
#We now had to clean up the Mission 1 column and split it again by the & sign, which was not needed.
new_2 = new_space_df['Mission part 1'].str.split("&", n=1, expand=True)

new_space_df['Mission 1']= new_2[0]
new_space_df['Mission 2']= new_2[1]

new_space_df.drop(columns= ['Mission part 1'], inplace= True)

new_space_df.head(50)
```

```python
#We had to do the same thing to Mission part 2 column - split it by the & sign, which was not needed.
new_3 = new_space_df['Mission part 2'].str.split("&", n=1, expand=True)

new_space_df['Mission 3']= new_3[0]
new_space_df['Mission 4']= new_3[1]

new_space_df.drop(columns= ['Mission part 2'], inplace= True)

new_space_df
```

```python
#Using the melt method, we stacked all the missions so there is one column for the missions.
transformed_space_df = pd.melt(new_space_df, id_vars = ['Company Name', 'Datum', 'Status Mission'],
                              value_vars = ['Mission 1', 'Mission 2', 'Mission 3', 'Mission 4'],
                              var_name = 'Mission_num', value_name = 'Missions')

transformed_space_df
```

```python
#We had to further clean the data and remove any spaces before or after the mission names so it does not skew
#the join below.
cleaned_space_df= transformed_space_df.dropna(axis=0, how='any', thresh=None, subset=None, inplace=False)
cleaned_space_df['Missions']= cleaned_space_df['Missions'].str.strip()
cleaned_space_df.head()
```

```python
#Here we did an inner join on "missions" so it joined both DF's on the Missions column.
astronauts_missions_df= pd.merge(cleaned_astro_df, cleaned_space_df, on= "Missions", how="inner")
astronauts_missions_df
```

```python
#We are connecting to our Database, which in this case is PostgreSQL.
rds_connection_string = (f"postgres:{password}@localhost:5432/etl_db")
engine = create_engine(f'postgresql://{rds_connection_string}')
```

```python
#Using pandas to load csv converted DataFrame into the SQL database above.
astronauts_missions_df.to_sql(name='astronaut_missions', con=engine, if_exists='append', index=False)
```

```python
#Confirming the data has been added by querying the astronaut_missions table.
pd.read_sql_query('SELECT * FROM astronaut_missions', con=engine).head()
```
