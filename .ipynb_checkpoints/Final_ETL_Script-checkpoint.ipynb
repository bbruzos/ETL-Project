```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
# Load in Astronaut Dataset
file_to_load= "csv_data/astronauts.csv"
astro_df= pd.read_csv(file_to_load)
astro_df.head()
```

```python
file_to_load= "csv_data/Space_Corrected.csv"
space_df= pd.read_csv(file_to_load)
space_df.head(50)
```

```python
new_astro_df= astro_df[['Name','Birth Date','Status','Gender','Missions']].copy()
new_astro_df.head()
```

```python
astro_data= new_astro_df['Missions'].str.split(",", n=3, expand=True)

new_astro_df['Mission 1']= astro_data[0]
new_astro_df['Mission 2']= astro_data[1]
new_astro_df['Mission 3']= astro_data[2]
new_astro_df['Mission 4']= astro_data[3]

new_astro_df.drop(columns= ['Missions'], inplace= True)

new_astro_df.head()
```

```python
new_astro_df['Mission 1']= new_astro_df['Mission 1'].str.split('(').str[0]
new_astro_df['Mission 2']= new_astro_df['Mission 2'].str.split('(').str[0]
new_astro_df['Mission 3']= new_astro_df['Mission 3'].str.split('(').str[0]
new_astro_df['Mission 4']= new_astro_df['Mission 4'].str.split('(').str[0]
new_astro_df
```

```python
transformed_astro_df= pd.melt(new_astro_df, id_vars =['Name','Birth Date','Status','Gender'], 
                            value_vars =['Mission 1','Mission 2','Mission 3','Mission 4'], 
                            var_name ='Mission_num', value_name ='Missions') 
transformed_astro_df
```

```python
cleaned_astro_df= transformed_astro_df.dropna(how='any') 
cleaned_astro_df.head()
```
